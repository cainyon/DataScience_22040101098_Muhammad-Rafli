```python
# Load Library dan Dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('titanic')

# Mengisi missing value pada umur dengan median agar grafik histogram/boxplot rapi
df['age'] = df['age'].fillna(df['age'].median())
```

```python
# Bar Chart : Total penumpang berdasarkan kelas
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='pclass', palette='Set2')
plt.title('Jumlah Penumpang Berdasarkan Kelas Tiket (Pclass)')
plt.xlabel('Kelas Tiket')
plt.ylabel('Jumlah Penumpang')
plt.show()

# Menghitung angka untuk narasi
print(df['pclass'].value_counts())
```

1. Grafik batang ini menunjukkan distribusi penumpang Titanic berdasarkan kelas tiket mereka, di mana Kelas 3 memiliki jumlah penumpang terbanyak yaitu 491 orang, diikuti Kelas 1 sebanyak 216 orang, dan Kelas 2 sebanyak 184 orang.
2. Data ini penting karena memperlihatkan bahwa mayoritas penumpang kapal (lebih dari 55%) berada di kelas ekonomi (Kelas 3), yang secara historis memiliki akses evakuasi paling terbatas menuju sekoci saat kapal tenggelam
3. Tindak lanjut yang bisa dilakukan adalah menganalisis apakah besarnya jumlah penumpang di Kelas 3 ini berbanding lurus dengan tingkat kematian mereka, atau apakah ada ketimpangan keselamatan yang nyata dibandingkan dengan Kelas 1 yang penumpangnya lebih sedikit

```python
# Histogram : Distribusi Umur Penumpang
plt.figure(figsize=(7, 4))
sns.histplot(data=df, x='age', bins=20, kde=True, color='skyblue')
plt.title('Distribusi Umur Penumpang Titanic')
plt.xlabel('Umur')
plt.ylabel('Frekuensi')
plt.show()

# Menghitung angka untuk narasi
print(df['age'].describe())
```

1. Histogram ini menunjukkan sebaran umur penumpang Titanic yang berpusat pada usia dewasa, dengan nilai tengah berada di angka 28 tahun dan rentang terbesar berada di antara usia 20 hingga 35 tahun.
2. Informasi ini krusial karena menunjukkan bahwa populasi kapal didominasi oleh orang dewasa muda, sementara kelompok rentan seperti bayi di bawah 5 tahun dan lansia di atas 60 tahun jumlahnya sangat sedikit.
3. Pertanyaan eksplorasi berikutnya adalah apakah prioritas evakuasi "women and children first" benar-benar berhasil menyelamatkan kelompok anak-anak di bawah umur tersebut, atau justru kelompok usia produktif yang mendominasi tingkat keselamatan karena kekuatan fisik mereka

```python
# Boxplot: distribusi umur berdasarkan kelas tiket
plt.figure(figsize=(7, 5))

# Memperbaiki peringatan dengan menambahkan hue='pclass' dan legend=False
sns.boxplot(data=df, x='pclass', y='age', hue='pclass', palette='Set3', legend=False)

plt.title('Distribusi Umur Penumpang Berdasarkan Kelas Tiket')
plt.xlabel('Kelas Tiket (Pclass)')
plt.ylabel('Umur Penumpang (Tahun)')
plt.show()
```

1. Boxplot menunjukkan distribusi umur penumpang di tiga kelas berbeda, di mana terdapat tren penurunan nilai tengah (median) umur seiring menurunnya kelas tiket; Kelas 1 memiliki median umur tertua yaitu 37 tahun, Kelas 2 sebesar 29 tahun, dan Kelas 3 didominasi oleh penumpang termuda dengan median 24 tahun.
2. Grafik memberikan konteks sosial yang penting bahwa penumpang yang mampu membeli fasilitas VIP (Kelas 1) rata-rata adalah kelompok usia matang yang mapan secara finansial.
3. Pertanyaan eksplorasi yang muncul adalah apakah faktor usia muda di Kelas 3 ini memberikan mereka keunggulan fisik untuk menyelamatkan diri saat kapal tenggelam, ataukah faktor usia matang di Kelas 1 tetap lebih diuntungkan karena prioritas fasilitas dan akses evakuasi.


```python
# Scatter Plot: hubungan umur dan tarif berdasarkan kelas
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='age', y='fare', hue='pclass', palette='bright', alpha=0.7)
plt.title('Hubungan Antara Umur, Tarif Tiket, dan Kelas Penumpang')
plt.xlabel('Umur')
plt.ylabel('Tarif Tiket (Fare)')
plt.ylim(0, 300) # Membatasi outlier fare agar sebaran titik terlihat baik
plt.show()
```

1. 2. Grafik memberikan konteks sosial yang penting bahwa penumpang yang mampu membeli fasilitas VIP (Kelas 1) rata-rata adalah kelompok usia matang yang mapan secara finansial.
